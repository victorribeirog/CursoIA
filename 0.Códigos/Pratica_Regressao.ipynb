```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import statsmodels.formula.api as smf
import scipy.stats as stats
import seaborn as sns
```

```python
base = pd.read_csv('mt_cars.csv')
base.shape
```

```python
base.head()
```

```python
base = base.drop(['Unnamed: 0'], axis=1)
base.head()
```

```python
corr = base.corr()
sns.heatmap(corr, cmap='coolwarm', annot=True, fmt='.2f')
```

```python
column_pairs = [('mpg', 'cyl'), ('mpg', 'disp'), ('mpg', 'hp'), ('mpg', 'drat'), ('mpg', 'wt'), ('mpg', 'vs')]
n_plots = len(column_pairs)
fig, axes = plt.subplots(nrows=n_plots, ncols=1, figsize=(6, 4 * n_plots))

for i, pair in enumerate(column_pairs):
    x_col, y_col = pair
    sns.scatterplot(data=base, x=x_col, y=y_col, ax=axes[i])
    axes[i].set_title(f'Scatter Plot: {x_col} vs {y_col}')

plt.tight_layout()
plt.show()
```

```python
# aic = 156.6 e bic = 162.5
#modelo = smf.ols(formula='mpg ~ wt + disp + hp', data=base).fit()

# aic = 165.1 e bic = 169.5
modelo = smf.ols(formula='mpg ~ disp + cyl', data=base).fit()

# aic = 179.1 e bic = 183.5
modelo = smf.ols(formula='mpg ~ drat + vs', data=base).fit()
print(modelo.summary())
```

```python
residuos = modelo.resid
plt.hist(residuos, bins=20, edgecolor='k')
plt.title('Histograma dos Resíduos')
plt.xlabel('Resíduos')
plt.ylabel('Frequência')
plt.show()
```

```python
stats.probplot(residuos, dist="norm", plot=plt)
plt.title('QQ Plot dos Resíduos')
plt.show()
```

```python
stats, pval = stats.shapiro(residuos)
print(f'Statistic: {stats}, p-value: {pval}')
if pval > 0.05:
    print("Os resíduos seguem uma distribuição normal.")
else:
    print("Os resíduos não seguem uma distribuição normal.")
```
